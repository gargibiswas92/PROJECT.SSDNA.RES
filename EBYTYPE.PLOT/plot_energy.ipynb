```python
import os
import matplotlib.pyplot as plt
from statistics import stdev
from statistics import mean
out1 = open("energy_stacking_200_07.txt", "w")

tt = []

for i in range(1, 50001, 1):
    tt.append(i)
    
mean_en = []
std_en = []

ss0 = ['Mean_energy  Stdev', "\n"]
out1.writelines(ss0)

for file in os.listdir():
    if file.startswith("EbyType") and file.endswith(".dat"):
        print(file)
        inp1 = open(file, "r")
        count = 0
        en = []
        for line in inp1:
            count = count + 1
            if count%2 ==1:
                arr = line.split()
                en.append(float(arr[11])) 
        plt.plot(tt, en)
        a = mean(en)
        b = stdev(en)
        ss = '{:>8.3f}{:>8.3f}{}'.format(a, b, "\n")
        out1.writelines(ss)
        mean_en.append(a)
        std_en.append(b)

plt.xlabel("Time steps")
plt.ylabel("$E_B$ [kcal/mol]")
plt.title("Protein-DNA Stacking, $K_H$ = 0.7")
plt.savefig("stacking_200_07.png", dpi=300)
print(mean(mean_en), mean(std_en))
out1.close()
inp1.close()
```
