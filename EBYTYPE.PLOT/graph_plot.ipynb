```python
import matplotlib.pyplot as plt
import numpy as np

y_sta_07 = [-14.87, -15.29, -14.98, -14.71, -14.38]
y_sta_00 = [-17.22, -16.15, -15.75, -14.77, -15.61]
y_sta_07_er = [2.3, 2.44, 2.40, 2.187, 2.007]
y_sta_00_er = [2.15, 2.16, 2.29, 1.95, 2.06]
barwidth = 0.25

x1 = np.arange(5)

plt.bar(x1+0.00, y_sta_07, color = 'magenta', width = 0.25, label = '$k_D$= 0.7')
plt.bar(x1+0.00, y_sta_07, yerr= y_sta_07_er, capsize= 6, color ='magenta', width = 0.25)
plt.bar(x1+0.25, y_sta_00, color = 'orange', width = 0.25, label = '$k_D$= 0.0')
plt.bar(x1+0.25, y_sta_00, yerr= y_sta_00_er, capsize=6, color ='orange', width = 0.25)
plt.xticks(x1, ['222', '420', '620', '812', '994'])
ax = plt.gca()
ax.invert_yaxis()

plt.xlabel("End-to-end distance")
plt.ylabel("Average Energy (kcal/mol)")
plt.title("Stacking Interaction between protein-DNA")
plt.ylim(-10, -20)
plt.legend()
plt.savefig("Stacking_075_compare.png", dpi=300)
```
