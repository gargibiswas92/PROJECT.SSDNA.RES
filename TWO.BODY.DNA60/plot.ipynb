```python
inp1 = open('average.txt','r')
inp2 = open('average1.txt','r')
index=[]
ene=[]
ene1=[]
for line in inp1:
    arr = line.split()
    index.append(int(arr[0]))
    ene.append(float(arr[1]))
    
    
for line2 in inp2:
    arr2 = line2.split()
    ene1.append(float(arr2[1]))
    
import matplotlib.pyplot as plt
plt.plot(index, ene, label='30 mM', color='black', linewidth=3, mec='black')
plt.plot(index, ene1, label='1000 mM', color='gray', linewidth=3)
plt.xlabel('Nucleotide Index', fontsize=18)
plt.ylabel('Interaction Energy\n(kcal/mol)', fontsize=18)
plt.legend(loc='upper center', fontsize=16)
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.savefig('energy_landscape.png', dpi=300, bbox_inches='tight')
```
