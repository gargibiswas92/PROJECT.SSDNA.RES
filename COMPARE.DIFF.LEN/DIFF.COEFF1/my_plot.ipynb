```python
import matplotlib.pyplot as plt
import numpy as np

# Load the data from files
dmod1 = np.loadtxt('diff_com_235.txt')
dmod2 = np.loadtxt('diff_com_460.txt')
dmod3 = np.loadtxt('diff_com_860.txt')
dmod4 = np.loadtxt('diff_com_1400.txt')


# Combine the data into a list
data_pm = [dmod1, dmod2, dmod3, dmod4]
data = np.multiply(data_pm, 15304)
# Create the figure and axes objects
fig, ax = plt.subplots()

# Create the boxplots with green and red boxes
bp = ax.boxplot(data, patch_artist=True, boxprops=dict(facecolor='coral'))


medians = [median.get_ydata() for median in bp['medians']]

ax.plot(np.arange(1, 5), medians, 'ko-', linewidth=1)
#print(medians, medians1)
xticklabels = ['238', '460', '860', '1400']
ax.set_xticks(np.arange(1, len(xticklabels) + 1))
ax.set_xticklabels(xticklabels, fontsize=14)
ax.set_ylabel('D in nt$^2$/timesteps', fontsize=18)
ax.set_xlabel('DNA length (nt)', fontsize=18)
#legend = ax.legend([bp["boxes"][0], bp1["boxes"][0]], ['10 mM', '300 mM'], loc='upper right')
```

```python
import matplotlib.pyplot as plt

means = []
stds = []

for item in data:
    mean_d = np.mean(item)
    std_d = np.std(item)
    means.append(mean_d)
    stds.append(std_d)
    

x = [235, 460, 860, 1400]
#plt.plot(x, means, 'o-', color='turquoise', markersize=6, edgecolors='black', label="low salt")
plt.errorbar(x, means, yerr=stds, capsize=6, fmt='o-', color= 'coral', markeredgecolor='black', ecolor='black', markersize=8)
#plt.plot(x, means1, 'o-', color='mediumorchid', markersize=6, edgecolors='black', label="high salt")
#plt.errorbar(x, means1, yerr=stds1, capsize=6, fmt='o-', color= 'mediumorchid', markeredgecolor='black', ecolor='black', markersize=8, label="300 mM")
#plt.legend(fontsize=16)
plt.ylabel('D in nt$^2$/timesteps', fontsize=18)
plt.xlabel('DNA length (nt)', fontsize=18)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.savefig('line_plot.png', dpi=300, bbox_inches='tight')
```
