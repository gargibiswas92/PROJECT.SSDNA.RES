```python
import matplotlib.pyplot as plt
inp1 = open('diff_07.txt', 'r')
inp2 = open('diff_00.txt', 'r')

dis = []
d = []
derr = []

d2 = []
derr2 = []

d_n = []
derr_n = []
d2_n = []
derr2_n = []

for line in inp1:
    arr = line.split()
    dis.append(int(arr[0]))
    d.append(float(arr[1]))
    derr.append(float(arr[2]))
    
for line2 in inp2:
    arr2 = line2.split()
    d2.append(float(arr2[1]))
    derr2.append(float(arr2[2]))
m =  1/(3.762184736801714e-07)
#m = 1/(2.1049946688170693e-08)
for i in range(len(d)):
    d_n.append((d[i]/2)*m)
    derr_n.append((derr[i]/2)*m)
    d2_n.append((d2[i]/2)*m)
    derr2_n.append((derr2[i]/2)*m)
    
plt.plot(dis, d_n, 'o-', color='teal', label='dihedral potential = 0.7')
plt.errorbar(dis, d_n, yerr=derr_n, capsize=6, color='teal')
plt.plot(dis, d2_n, 'o-', color='mediumorchid', label= 'dihedral potential = 0.0')
plt.errorbar(dis, d2_n, yerr=derr2_n, capsize=6, color='mediumorchid')
plt.legend()
import numpy as np
plt.xlabel('End-to-end distance ($\AA$)', fontsize=14)
plt.ylabel('Diffusion Coefficient (nt$^2$/s)', fontsize=14)
#plt.yticks(np.arange(0, 370000, step=50000), ['0', '5x10$^4$', '10x10$^4$', '15x10$^4$', '20x10$^4$', '25x10$^4$', '30x10$^4$', '35x10$^4$'])
plt.savefig('figure_N.png', dpi=300, bbox_inches='tight')

out1 = open('new_diff.txt', 'w')
sp = ['Distance         dh=0.7             dh=0.0       ','\n']
out1.writelines(sp)
for j in range(len(d_n)):
    ss='{:>8d}{}{:>6.3e}{}{:>6.3e}{}{:>6.3e}{}{:>6.3e}{}'.format(dis[j],'  ', d_n[j], '  ', derr_n[j], '  ',d2_n[j], '  ', derr2_n[j],'\n')
    ss='{:>8d}{}{:>10.1f}{}{:>10.1f}{}{:>10.1f}{}{:>10.1f}{}'.format(dis[j],'  ', d_n[j], '  ', derr_n[j], '  ',d2_n[j], '  ', derr2_n[j],'\n')
    out1.writelines(ss)
out1.close()    
```
