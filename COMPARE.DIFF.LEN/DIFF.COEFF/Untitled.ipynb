```python
import matplotlib.pyplot as plt
import numpy as np

x2 = [235, 460, 868]
y2 = [0.153, 1.052, 1.826]
y_err2 = [0.101, 0.989, 2.220]

x1 = []
y1 = []
y_err = []
m =  1/(6.272365708359268e-08)
for i in range(len(x2)):
    a = y2[i]*m
    b = y_err2[i]*m
    y1.append(a)
    y_err.append(b)

plt.plot(x2, y1, 'o-', color='darkcyan')
#plt.plot(x, y, 'o')
plt.errorbar(x2, y1, yerr=y_err, capsize=6, color='darkcyan')
plt.xlabel('DNA length in nt', fontsize=16, fontweight='bold', family='serif')
plt.ylabel('[D] in nt$^{2}$/s', fontsize=16, fontweight='bold', family='serif')
plt.yticks(np.arange(0, 70000000, step=10000000), ['0', '10x10$^6$', '20x10$^6$', '30x10$^6$', '40x10$^6$', '50x10$^6$', '60x10$^6$'])
plt.xticks(fontsize=12, weight='bold', family='serif')
plt.yticks(fontsize=12, weight='bold', family='serif')
plt.title('Diffusion at different DNA length', fontsize=14, weight='bold', family='serif')
plt.savefig('Diff_len2.png', dpi=300, bbox_inches='tight')
```
