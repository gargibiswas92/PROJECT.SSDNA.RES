```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import gaussian_kde
import seaborn as sns

jump1 = []
with open('all_int_trans_200.txt', 'r') as inp1:
    for line in inp1:
        arr = line.split()
        a = float(arr[1])
        if a >= 10.0:
            jump1.append(a)
            
jump2 = []
with open('all_int_trans_460.txt', 'r') as inp2:
    for line2 in inp2:
        arr2 = line2.split()
        a2 = float(arr2[1])
        if a2 >= 10.0:
            jump2.append(a2)
            
jump3 = []
with open('all_int_trans_860.txt', 'r') as inp3:
    for line3 in inp3:
        arr3 = line3.split()
        a3 = float(arr3[1])
        if a3 >= 10.0:
            jump3.append(a3)
            
jump4 = []
with open('all_int_trans_1400.txt', 'r') as inp4:
    for line4 in inp4:
        arr4 = line4.split()
        a4 = float(arr4[1])
        if a4 >= 10.0:
            jump4.append(a4)
data1 = np.array(jump1)
data2 = np.array(jump2)
data3 = np.array(jump3)
data4 = np.array(jump4)
   
data = [data1, data2, data3, data4]

bins = np.linspace(0, 350, 70)

counts1, bins1, _ = plt.hist(jump1, bins=bins, range=(0, 350))
counts2, bins2, _ = plt.hist(jump2, bins=bins, range=(0, 350))
counts3, bins3, _ = plt.hist(jump3, bins=bins, range=(0, 350))
counts4, bins4, _ = plt.hist(jump4, bins=bins, range=(0, 350))


plt.clf()

constant1 = 7.0

normalized_counts1 = counts1 / constant1
#pdf_norm1 = pdf1 / constant1

constant2 = 10.0

normalized_counts2 = counts2 / constant2
#pdf_norm2 = pdf2 / constant2

normalized_counts3 = counts3 / constant2
#pdf_norm3 = pdf3 / constant2

constant3 = 6.0

normalized_counts4 = counts4 / constant3
#pdf_norm4 = pdf4 / constant3

fig, ax = plt.subplots(nrows=1, ncols=4, figsize=(25,6))

num_colors = 5
colors = plt.cm.Set3(np.linspace(0, 1, num_colors))

ax[0].bar(bins1[:-1], normalized_counts1, width=(bins[1]-bins[0]), color=colors[0], label='235 NT')
ax[1].bar(bins2[:-1], normalized_counts2, width=(bins[1]-bins[0]), color=colors[1], label='460 NT')
ax[2].bar(bins3[:-1], normalized_counts3, width=(bins[1]-bins[0]), color=colors[2], label='860 NT')
ax[3].bar(bins4[:-1], normalized_counts4, width=(bins[1]-bins[0]), color=colors[3], label='1400 NT')
#ax[4].bar(bins5[:-1], normalized_counts5, width=(bins[1]-bins[0]), color=colors[4])

#ax.set_xticklabels(ax.get_xticklabels(), fontsize=14)
ax[0].set_ylabel('Frequency',fontsize=24)
ax[1].set_xlabel('Size of Intersegmental Transfer (ΔNT)',fontsize=24)
#ax[0].set_ylim(0,145)
ax[2].set_ylim(0,85)
ax[3].set_ylim(0,85)
ax[0].tick_params(labelsize=18)
ax[1].tick_params(labelsize=18)
ax[2].tick_params(labelsize=18)
ax[3].tick_params(labelsize=18)
ax[0].legend(fontsize=20)
ax[1].legend(fontsize=20)
ax[2].legend(fontsize=20)
ax[3].legend(fontsize=20)
#ax[4].tick_params(labelsize=18)
#ax.set_xticklabels(ax.get_xticklabels(), fontsize=16)
plt.savefig('diffdis_freq.png', dpi=300, bbox_inches='tight')
```
