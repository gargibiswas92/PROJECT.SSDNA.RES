```python
inp1 = open('all_int_trans.txt','r')
jump = []
for line in inp1:
    arr = line.split()
    a = float(arr[1])
    if a >= 10.0:
        jump.append(a)
    else:
        continue
import matplotlib.pyplot as plt
import numpy as np

data = np.array(jump)

counts, bins, _ = plt.hist(jump, bins=50, range=(0,200))

constant = 10.0
normalized_counts = counts / constant

plt.bar(bins[:-1], normalized_counts, width=(bins[1]-bins[0]))
plt.xlabel('ΔNT between 1000 consecutive MD steps', fontsize=16)
plt.ylabel('Frequency', fontsize=16)
#plt.title('Histogram with Frequency Normalized by 2')

    
```
