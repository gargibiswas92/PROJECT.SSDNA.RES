```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import gaussian_kde

jump1 = []
with open('all_int_trans_200_0.txt', 'r') as inp1:
    for line in inp1:
        arr = line.split()
        a = float(arr[1])
        if a >= 10.0:
            jump1.append(a)
            
jump2 = []
with open('all_int_trans_400_0.txt', 'r') as inp2:
    for line2 in inp2:
        arr2 = line2.split()
        a2 = float(arr2[1])
        if a2 >= 10.0:
            jump2.append(a2)
            
jump3 = []
with open('all_int_trans_600_0.txt', 'r') as inp3:
    for line3 in inp3:
        arr3 = line3.split()
        a3 = float(arr3[1])
        if a3 >= 10.0:
            jump3.append(a3)
            
jump4 = []
with open('all_int_trans_800_0.txt', 'r') as inp4:
    for line4 in inp4:
        arr4 = line4.split()
        a4 = float(arr4[1])
        if a4 >= 10.0:
            jump4.append(a4)
            
jump5 = []
with open('all_int_trans_1000_0.txt', 'r') as inp5:
    for line5 in inp5:
        arr5 = line5.split()
        a5 = float(arr5[1])
        if a5 >= 10.0:
            jump5.append(a5)



bins = np.linspace(0, 200, 50)

counts1, bins1, _ = plt.hist(jump1, bins=bins, range=(0, 200))
counts2, bins2, _ = plt.hist(jump2, bins=bins, range=(0, 200))
counts3, bins3, _ = plt.hist(jump3, bins=bins, range=(0, 200))
counts4, bins4, _ = plt.hist(jump4, bins=bins, range=(0, 200))
counts5, bins5, _ = plt.hist(jump5, bins=bins, range=(0, 200))

plt.clf()

constant1 = 10.0
constant2 = 9.0
constant3 = 7.0

normalized_counts1 = counts1 / constant1
normalized_counts2 = counts2 / constant1
normalized_counts3 = counts3 / constant1
normalized_counts4 = counts4 / constant1
normalized_counts5 = counts5 / constant1

fig, ax = plt.subplots(nrows=1, ncols=5, figsize=(25,6))

num_colors = 5
colors = plt.cm.rainbow(np.linspace(0, 1, num_colors))

ax[0].bar(bins1[:-1], normalized_counts1, width=(bins[1]-bins[0]), color=colors[0], label= '222$\AA$')
ax[1].bar(bins2[:-1], normalized_counts2, width=(bins[1]-bins[0]), color=colors[1], label= '420$\AA$')
ax[2].bar(bins3[:-1], normalized_counts3, width=(bins[1]-bins[0]), color=colors[2], label= '620$\AA$')
ax[3].bar(bins4[:-1], normalized_counts4, width=(bins[1]-bins[0]), color=colors[3], label= '812$\AA$')
ax[4].bar(bins5[:-1], normalized_counts5, width=(bins[1]-bins[0]), color=colors[4], label= '994$\AA$')

#ax.set_xticklabels(ax.get_xticklabels(), fontsize=14)
ax[0].set_ylabel('Frequency',fontsize=24)
ax[2].set_xlabel('Size of Intersegmental Transfer (Î”nt)',fontsize=24)
ax[0].set_ylim(0,145)
ax[1].set_ylim(0,145)
ax[4].set_ylim(0,145)
ax[0].tick_params(labelsize=18)
ax[1].tick_params(labelsize=18)
ax[2].tick_params(labelsize=18)
ax[3].tick_params(labelsize=18)
ax[4].tick_params(labelsize=18)

ax[0].legend(fontsize=20)
ax[1].legend(fontsize=20)
ax[2].legend(fontsize=20)
ax[3].legend(fontsize=20)
ax[4].legend(fontsize=20)
#ax.set_xticklabels(ax.get_xticklabels(), fontsize=16)
plt.savefig('highsalt_freq.png', dpi=300, bbox_inches='tight')
```
