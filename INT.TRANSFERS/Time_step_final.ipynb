```python
import numpy as np
import matplotlib.pyplot as plt

d1 = np.loadtxt('all_int_trans_200_7.txt')
d2 = np.loadtxt('all_int_trans_400_7.txt')
d3 = np.loadtxt('all_int_trans_600_7.txt')
d4 = np.loadtxt('all_int_trans_800_7.txt')
d5 = np.loadtxt('all_int_trans_1000_7.txt')

time1 = d1[:,0]
time2 = d2[:,0]
time3 = d3[:,0]
time4 = d4[:,0]
time5 = d5[:,0]

jump1 = d1[:,1]
jump2 = d2[:,1]
jump3 = d3[:,1]
jump4 = d4[:,1]
jump5 = d5[:,1]

x1 = len(jump1) // 49998
x2 = len(jump2) // 49998
x3 = len(jump3) // 49998
x4 = len(jump4) // 49998
x5 = len(jump5) // 49998

jump1_ind = np.array_split(jump1, x1)
time1_ind = np.array_split(time1, x1)

jump2_ind = np.array_split(jump2, x2)
time2_ind = np.array_split(time2, x2)

jump3_ind = np.array_split(jump3, x3)
time3_ind = np.array_split(time3, x3)

jump4_ind = np.array_split(jump4, x4)
time4_ind = np.array_split(time4, x4)

jump5_ind = np.array_split(jump5, x5)
time5_ind = np.array_split(time5, x5)
```

```python
import statistics as stat

def difference(arr_time):
      diff_time = []
      if len(arr_time) > 2:
         for i in range(len(arr_time)-1):
            diff_i = (arr_time[i+1] - arr_time[i])
            diff_time.append(diff_i)
         diff_time_np = np.array(diff_time)
         avg_time = stat.mean(diff_time_np)*(3.267)*(10**(-5))
         stddev_time = stat.stdev(diff_time_np)*(3.267)*(10**(-5))
      else:
         avg_time = 0.0
         stddev_time = 0.0
      #return avg_time, stddev_time
      return avg_time
```

```python
def time_step_calc(x1, jump1_ind, time1_ind):
  
   arr0_time = []
   arr1_time = []
   arr2_time = []
   arr3_time = []
   arr4_time = []
   arr5_time = []
   arr6_time = []
   arr7_time = []
   arr8_time = []
   arr9_time = []

   for k1 in range(x1):
      for item, item1 in zip(jump1_ind[k1], time1_ind[k1]):
         if item <= 50.0:
            arr0_time.append(item1)
         if item >= 50.0 and item <= 75.0:
            arr1_time.append(item1)
         if item >= 75.0 and item <= 100.0:
            arr2_time.append(item1)
         if item >= 100.0 and item <= 125.0:
            arr3_time.append(item1)
         if item >= 125.0 and item <= 150.0:
            arr4_time.append(item1)
         if item >= 150.0 and item <= 175.0:
            arr5_time.append(item1)
         if item >= 175.0 and item <= 200.0:
            arr6_time.append(item1)
         if item >= 200.0 and item <= 225.0:
            arr7_time.append(item1)
         if item >= 225.0 and item <= 250.0:
            arr8_time.append(item1)
         if item >= 250.0:
            arr9_time.append(item1)

   new0_time = np.array(arr0_time)
   new1_time = np.array(arr1_time)
   new2_time = np.array(arr2_time)
   new3_time = np.array(arr3_time)
   new4_time = np.array(arr4_time)
   new5_time = np.array(arr5_time)
   new6_time = np.array(arr6_time)
   new7_time = np.array(arr7_time)
   new8_time = np.array(arr8_time)
   new9_time = np.array(arr9_time)

   a0 = difference(new0_time)
   a1 = difference(new1_time)
   a2 = difference(new2_time)
   a3 = difference(new3_time)
   a4 = difference(new4_time)
   a5 = difference(new5_time)
   a6 = difference(new6_time)
   a7 = difference(new7_time)
   a8 = difference(new8_time)
   a9 = difference(new9_time)

   return a0, a1, a2, a3, a4, a5, a6, a7, a8, a9
```

```python
n1 = time_step_calc(x1, jump1_ind, time1_ind)
n2 = time_step_calc(x2, jump2_ind, time2_ind)
n3 = time_step_calc(x3, jump3_ind, time3_ind)
n4 = time_step_calc(x4, jump4_ind, time4_ind)
n5 = time_step_calc(x5, jump5_ind, time5_ind)

list1 = list(n1)
list2 = list(n2)
list3 = list(n3)
list4 = list(n4)
list5 = list(n5)

title = ['<50', '50-75', '75-100', '100-125', '125-150', '150-175', '175-200', '200-225', '225-250', '>250']

import pandas as pd

data1 = {'Size of Intersegmental \n Transfer (nt)': title, '222 $\AA$': list1, '420 $\AA$': list2, '620 $\AA$': list3, '812 $\AA$': list4,'994 $\AA$': list5}

data = pd.DataFrame(data1)

fig, ax = plt.subplots()

table = ax.table(cellText=data.values, colLabels=data.columns, loc='center')

table.auto_set_font_size(False)
table.set_fontsize(20)
table.scale(10, 4)
ax.axis('off')
fig.savefig('table_07.png', dpi= 300, bbox_inches='tight')
```
