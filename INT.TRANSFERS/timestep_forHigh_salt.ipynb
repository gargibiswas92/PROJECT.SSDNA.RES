```python
import numpy as np
import matplotlib.pyplot as plt

d1 = np.loadtxt('all_int_trans_200_0.txt')
d2 = np.loadtxt('all_int_trans_400_0.txt')
d3 = np.loadtxt('all_int_trans_600_0.txt')
d4 = np.loadtxt('all_int_trans_800_0.txt')
d5 = np.loadtxt('all_int_trans_1000_7.txt')

time1 = d1[:,0]
time2 = d2[:,0]
time3 = d3[:,0]
time4 = d4[:,0]
time5 = d5[:,0]

jump1 = d1[:,1]
jump2 = d2[:,1]
jump3 = d3[:,1]
jump4 = d4[:,1]
jump5 = d5[:,1]

x1 = len(jump1) // 49998
x2 = len(jump2) // 49998
x3 = len(jump3) // 49998
x4 = len(jump4) // 49998
x5 = len(jump5) // 49998

jump1_ind = np.array_split(jump1, x1)
time1_ind = np.array_split(time1, x1)

jump2_ind = np.array_split(jump2, x2)
time2_ind = np.array_split(time2, x2)

jump3_ind = np.array_split(jump3, x3)
time3_ind = np.array_split(time3, x3)

jump4_ind = np.array_split(jump4, x4)
time4_ind = np.array_split(time4, x4)

jump5_ind = np.array_split(jump5, x5)
time5_ind = np.array_split(time5, x5)
```

```python
def int_binning(jump):
    bin1 = []
    bin2 = []
    bin3 = []
    bin4 = []
    bin5 = []
    bin6 = []
    bin7 = []
    bin8 = []
    bin9 = []
    bin10 = []
    for item in jump:
        if item <= 50.0 and item >=20.0:
            bin1.append(item)
        if item >= 50.0 and item <= 75.0:
            bin2.append(item)
        if item >= 75.0 and item <= 100.0:
            bin3.append(item)
        if item >= 100.0 and item <= 125.0:
            bin4.append(item)
        if item >= 125.0 and item <= 150.0:
            bin5.append(item)
        if item >= 150.0 and item <= 175.0:
            bin6.append(item)
        if item >= 175.0 and item <= 200.0:
            bin7.append(item)
        if item >= 200.0 and item <= 225.0:
            bin8.append(item)
        if item >= 225.0 and item <= 250.0:
            bin9.append(item)
        if item >= 250.0:
            bin10.append(item)
    return len(bin1), len(bin2), len(bin3), len(bin4), len(bin5), len(bin6), len(bin7), len(bin8), len(bin9), len(bin10)

jump1_bin = int_binning(jump1)
jump2_bin = int_binning(jump2)
jump3_bin = int_binning(jump3)
jump4_bin = int_binning(jump4)
jump5_bin = int_binning(jump5)
```

```python
print(jump1_bin)
print(jump2_bin)
print(jump3_bin)
print(jump4_bin)
print(jump5_bin)

def find_average(bin, x):
    av_list = []
    for item in bin:
        a = item/x
        av_list.append(a)
    return av_list
av_bin1 = find_average(jump1_bin, x1)
av_bin2 = find_average(jump2_bin, x2)
av_bin3 = find_average(jump3_bin, x3)
av_bin4 = find_average(jump4_bin, x4)
av_bin5 = find_average(jump5_bin, x5)

print(av_bin1)
print(av_bin2)
print(av_bin3)
print(av_bin4)
print(av_bin5)

def find_percentage(jump_bin):
    sum = 0.0
    perc_bin = []
    for item in jump_bin:
        sum= sum+item
    for item in jump_bin:
        perc = (item/sum)*100
        perc_bin.append(perc)
    return perc_bin

perc_bin1 = find_percentage(av_bin1)
perc_bin2 = find_percentage(av_bin2)
perc_bin3 = find_percentage(av_bin3)
perc_bin4 = find_percentage(av_bin4)
perc_bin5 = find_percentage(av_bin5)
```
