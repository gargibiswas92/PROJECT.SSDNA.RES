```python
import matplotlib.pyplot as plt
import numpy as np

# Load the data from files
dmod1 = np.loadtxt('diff_com_200_7.txt')
dmod2 = np.loadtxt('diff_com_400_7.txt')
dmod3 = np.loadtxt('diff_com_600_7.txt')
dmod4 = np.loadtxt('diff_com_800_7.txt')
dmod5 = np.loadtxt('diff_com_1000_7.txt')

dmop1 = np.loadtxt('diff_com_200_0.txt')
dmop2 = np.loadtxt('diff_com_400_0.txt')
dmop3 = np.loadtxt('diff_com_600_0.txt')
dmop4 = np.loadtxt('diff_com_800_0.txt')
dmop5 = np.loadtxt('diff_com_1000_0.txt')

# Combine the data into a list
data = [dmod1, dmod2, dmod3, dmod4, dmod5]
data1 = [dmop1, dmop2, dmop3, dmop4, dmop5]

# Create the figure and axes objects
fig, ax = plt.subplots()

# Create the boxplots with green and red boxes
bp = ax.boxplot(data, patch_artist=True, boxprops=dict(facecolor='turquoise'))
bp1 = ax.boxplot(data1, patch_artist=True, boxprops=dict(facecolor='mediumorchid'))



medians = [median.get_ydata() for median in bp['medians']]
medians1 = [median.get_ydata() for median in bp1['medians']]
ax.plot(np.arange(1, 6), medians, 'ko-', linewidth=1)
ax.plot(np.arange(1, 6), medians1, 'ko-', linewidth=1)
#print(medians, medians1)
xticklabels = ['222', '420', '620', '812', '994']
ax.set_xticks(np.arange(1, len(xticklabels) + 1))
ax.set_xticklabels(xticklabels, fontsize=14)
ax.set_ylabel('D in nt$^2$/timesteps', fontsize=18)
ax.set_xlabel('End-to-end distance (Å)', fontsize=18)
legend = ax.legend([bp["boxes"][0], bp1["boxes"][0]], ['10 mM', '300 mM'], loc='upper right')
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Load the data from files
dmod1 = np.loadtxt('diff_com_200_7.txt')
dmod2 = np.loadtxt('diff_com_400_7.txt')
dmod3 = np.loadtxt('diff_com_600_7.txt')
dmod4 = np.loadtxt('diff_com_800_7.txt')
dmod5 = np.loadtxt('diff_com_1000_7.txt')

dmop1 = np.loadtxt('diff_com_200_0.txt')
dmop2 = np.loadtxt('diff_com_400_0.txt')
dmop3 = np.loadtxt('diff_com_600_0.txt')
dmop4 = np.loadtxt('diff_com_800_0.txt')
dmop5 = np.loadtxt('diff_com_1000_0.txt')

# Combine the data into a list
data_pm = [dmod1, dmod2, dmod3, dmod4, dmod5]
data1_pm = [dmop1, dmop2, dmop3, dmop4, dmop5]


data = np.multiply(data_pm, 15304)
data1 = np.multiply(data1_pm, 15304)
# Create the figure and axes objects

ig, ax = plt.subplots()

# Create the boxplots with green and red boxes
bp = ax.boxplot(data, patch_artist=True, boxprops=dict(facecolor='turquoise'))
bp1 = ax.boxplot(data1, patch_artist=True, boxprops=dict(facecolor='mediumorchid'))



medians = [median.get_ydata() for median in bp['medians']]
medians1 = [median.get_ydata() for median in bp1['medians']]
ax.plot(np.arange(1, 6), medians, 'ko-', linewidth=1)
ax.plot(np.arange(1, 6), medians1, 'ko-', linewidth=1)
#print(medians, medians1)
xticklabels = ['222', '420', '620', '812', '994']
ax.set_xticks(np.arange(1, len(xticklabels) + 1))
ax.set_xticklabels(xticklabels, fontsize=14)
#ax.set_yticklabels(fontsize=14)
ax.set_ylabel('D in nt$^2$/timesteps', fontsize=18)
ax.set_xlabel('End-to-end distance (Å)', fontsize=18)
legend = ax.legend([bp["boxes"][0], bp1["boxes"][0]], ['10 mM', '300 mM'], loc='upper right', fontsize=16)
plt.yticks(fontsize=14)
plt.savefig('box_plot.png', dpi=300, bbox_inches='tight')
```

```python
import matplotlib.pyplot as plt

means = []
stds = []
means1 = []
stds1 = []

for item in data:
    mean_d = np.mean(item)
    std_d = np.std(item)
    means.append(mean_d)
    stds.append(std_d)
    
for item1 in data1:
    mean_d = np.mean(item1)
    std_d = np.std(item1)
    means1.append(mean_d)
    stds1.append(std_d)

x = [222.0, 420.0, 620.0, 812.0, 994.0]
#plt.plot(x, means, 'o-', color='turquoise', markersize=6, edgecolors='black', label="low salt")
plt.errorbar(x, means, yerr=stds, capsize=6, fmt='o-', color= 'turquoise', markeredgecolor='black', ecolor='black', markersize=8, label="10 mM")
#plt.plot(x, means1, 'o-', color='mediumorchid', markersize=6, edgecolors='black', label="high salt")
plt.errorbar(x, means1, yerr=stds1, capsize=6, fmt='o-', color= 'mediumorchid', markeredgecolor='black', ecolor='black', markersize=8, label="300 mM")
plt.legend(fontsize=16)
plt.ylabel('D in nt$^2$/timesteps', fontsize=18)
plt.xlabel('End-to-end distance (Å)', fontsize=18)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.savefig('line_plot.png', dpi=300, bbox_inches='tight')
```
